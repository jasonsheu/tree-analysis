```python
trees = pd.read_csv('corrected_diameter.csv')
parks = pd.read_csv('Tree Mapping Sign Up Sheet.csv')

parks.columns = ['Park Name', 'Student name', 'Email']
percentages = (trees['Fruit/Seed'].value_counts() / len(trees)) * 100
percentages.name = 'Percentage of total'
#pd.DataFrame(percentages)
```

```python
def clean_str(str):
    return str.lower()
trees['Email'] = trees['Email'].astype(str)
trees['Email'] = trees['Email'].apply(clean_str)
parks['Email'] = parks['Email'].astype(str)
parks['Email'] = parks['Email'].apply(clean_str)
parks = parks.drop_duplicates(subset=['Email'])
parks = parks[['Park Name', 'Email']]
```

```python
parks
```

```python

trees=trees.merge(parks, on='Email', how='left')
trees.to_csv('trees with park name.csv')
```
